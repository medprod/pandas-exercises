### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
#change data type of item_price from object to float
fxn = lambda row: float(row[1:-1])
chipo['item_price'] = chipo['item_price'].apply(fxn)
```

```python
#dropping any duplicate values with all 3 columns having same values
chipo_drop = chipo.drop_duplicates(['quantity', 'item_name', 'choice_description'])
#choose products > 10 and have a quantity of 1
products = chipo_drop[(chipo_drop['item_price']>10.00) & (chipo_drop['quantity']==1)]
#counting unique item names from products df
products['item_name'].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_unique = chipo.drop_duplicates(['item_name'])
price = chipo_unique.groupby('item_name')['item_price'].max()
price_df = price.reset_index()
price_df
```

### Step 6. Sort by the name of the item

```python
sorted_df = price_df['item_name'].sort_values()
sorted_df.reset_index()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = chipo.groupby('item_name')[['quantity', 'item_price']].max()
most_expensive_item['quantity'].sort_values(ascending=False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie = chipo[chipo['item_name']=='Veggie Salad Bowl'][['quantity', 'item_name', 'item_price']]
veggie['quantity'].sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda = chipo[chipo['item_name']=='Canned Soda'][['quantity', 'item_name', 'item_price']]
times = canned_soda[canned_soda['quantity']>1]
times['quantity'].count()
```
